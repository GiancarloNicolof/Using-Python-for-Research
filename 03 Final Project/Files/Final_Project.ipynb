```python
import numpy as np
import pandas as pd
```

Opening the training data, dropping unused columns (Unnamed, accuracy), setting the "timestamp" column as index.

```python
training_data = pd.read_csv("train_time_series.csv")
training_data = training_data.drop(training_data.columns[0], axis=1)
training_data = training_data.drop("accuracy", axis=1)
training_data = training_data.set_index("timestamp")

labels = pd.read_csv("train_labels.csv")
labels = labels.drop(labels.columns[0], axis=1)
labels = labels.set_index("timestamp")

print("Raw accelerometer data:")
print(training_data.head())

print("\nActivity labels:")
print(labels.head())
```

Synced the two dataframes into one.

```python
training_data["label"] = 0

for i in labels.index:
    training_data.loc[i, "label"] = labels.loc[i, "label"]

training_data.head(20)
```

Dropped the first 4 lines since the data in incomplete.

```python
training_data = training_data.drop(training_data.iloc[:4].index)
training_data.head(20)
```
